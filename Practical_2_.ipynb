```python
#How to compute averages using a sliding window over an array?
def moving_average(a,n=3):
  ret=np.cumsum(a,dtype=float)
  ret[n:]= ret[n:]-ret[:-n]
  return ret[n-1:]/n
a=np.arange(20)
print(moving_average(a,n=3))
```

```python
#Distance from a point to each line
def distance(P0,P1,p): T=P1-P0; L=(T**2).sum(axis=1); U=((P0[:,0]-p[...,0])*T[:,0] + (P0[:,1]-p[...,1])*T[:,1])/L; U=U.reshape(len(U),1); D=P0+U*T-p; return np.sqrt((D**2).sum(axis=1))
```

```python
#Consider a 16x16 array, how to get the block-sum (block size is 4x4)?
Z = np.ones((16,16)); k=4; S=np.add.reduceat(np.add.reduceat(Z, np.arange(0,Z.shape[0],k),axis=0), np.arange(0,Z.shape[1],k),axis=1); print(S)
```
