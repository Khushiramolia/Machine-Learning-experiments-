```python
#Importing data set from URLs
# Read CSV file from URL with default settings
url='https://tinyurl.com/yx3b6sq3'
data=pd.read_csv(url)
print(data)
```

```python
#Using different separator from file like ; or other
# Read CSV file with semicolon as separator
data=pd.read_csv(url,sep=';')
print(data)
```

```python
#Using column names explicitly
# Read CSV file and assign custom column names
data=pd.read_csv(url,sep=';',names=['CIC0','SM1_Dz(Z)','GATS1i','NdsCH'
,'NdssC','MLOGP','quantitative response LC50'])
print(data)
```

```python
#Using any column as an index
# Read CSV file and set column 1 as index
data=pd.read_csv(url,sep=';',index_col=1)
print(data)
```

```python
#Getting selected columns data from CSV file
# Read only selected columns (by index) from CSV
data=pd.read_csv(url,sep=';' ,usecols=[1,3,4])
print(data)
```

```python
#Prefix to add to column numbers when no header is available
# Read CSV with no header and assign custom names with prefix
data = pd.read_csv(url, sep=';', header=None, names=[f'Column{i}' for i in range(7)])
print(data)
```

```python
#Skipping selected rows (callable list)
data=pd.read_csv(url,sep=';',usecols=[1,5,6],skiprows=lambda x:
x%2==0)
print(data)
```
