```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# CGPA data for different semesters
cgpa1=[9.2,6.2,8.6,8.2,6.7,5.4,6,8]
cgpa2=[7,7.7,6.7,9,8,5.6,8,6]
cgpa3=[7.2,6.7,7.7,7,6,6.6,7,5.4]
cgpa4=[7.6,8.8,8.7,9,8,7.6,9,9.4]
sem=[1,2,3,4,5,6,7,8]
```

```python
# Create a new figure with specific size
plt.figure(figsize=(8, 4))
# Line plot of sem vs cgpa1 with markers
plt.plot(sem, cgpa1, marker='o', label='CGPA1')
# Scatter plot of sem vs cgpa2
plt.scatter(sem, cgpa2, marker='o', label='CGPA2 (points)')
# Add axis labels and title
plt.xlabel('Semester')
plt.ylabel('CGPA')
plt.title('CGPA Over Semesters')
# Show legend
plt.legend()
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Another line and scatter plot example
# cgpa1=[9.2,6.2,8.6,8.2,6.7,5.4,6,8]
# cgpa2=[7,7.7,6.7,9,8,5.6,8,6]
# cgpa3=[7.2,6.7,7.7,7,6,6.6,7,5.4]
# cgpa4=[7.6,8.8,8.7,9,8,7.6,9,9.4]
# sem=[1,2,3,4,5,6,7,8]
plt.plot(sem, cgpa1, label='CGPA1')        # Line plot
plt.scatter(sem, cgpa2, label='CGPA2 (points)')     # Scatter points
plt.xlabel('Semester')
plt.ylabel('CGPA')
plt.title('CGPA Over Semesters')
plt.ylim(0, 10) # Limit y-axis range
plt.legend()
plt.show()
```

```python
# Convert sem to numpy array for bar positions
# Prepare indices for grouped bar chart
sem_index = np.arange(len(sem))
bar_width = 0.2

# Plot grouped bars
plt.bar(sem_index - bar_width, cgpa1, width=bar_width)
plt.bar(sem_index, cgpa2, width=bar_width)
plt.bar(sem_index + bar_width, cgpa3, width=bar_width)

plt.ylim(0, 10)  # match y-axis from example
plt.xlabel("Semester")
plt.ylabel("CGPA")
plt.legend(['CGPA1', 'CGPA2', 'CGPA3'])
plt.title("Semester VS CGPA")
plt.xticks(sem_index, sem)

plt.show()
```

```python
# Multiple line plots on same graph
plt.plot(sem, cgpa1, label="L1")
plt.plot(sem, cgpa2, label="L2")
plt.plot(sem, cgpa3, label="L3")

plt.ylim(0, 10)
plt.xlabel("Semester")
plt.ylabel("CGPA")
plt.title("Semester VS CGPA")
plt.legend()
plt.show()
```

```python
# Subplots in 1 row, 2 columns
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(sem, cgpa1)
plt.legend(['CGPA1'])

plt.subplot(1, 2, 2)
plt.plot(sem, cgpa2)
plt.legend(['CGPA2'])

plt.show()
```

```python
# Subplots in 2 rows, 2 columns
plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.plot(sem, cgpa1)
plt.legend(['CGPA1'])

plt.subplot(2, 2, 2)
plt.plot(sem, cgpa2)
plt.legend(['CGPA2'])

plt.subplot(2, 2, 3)
plt.bar(sem, cgpa3)
plt.legend(['CGPA3'])

plt.subplot(2, 2, 4)
plt.bar(sem, cgpa4, width=0.3)
plt.legend(['CGPA4'])

plt.show()
```

```python
# Subplots with titles and labels
plt.figure(figsize=(10, 6))

# Top-left
plt.subplot(2, 2, 1)
plt.plot(sem, cgpa1, marker='o')
plt.title('Year VS Price')
plt.legend(['CGPA1'])
plt.xlabel('Year')
plt.ylabel('Price')

# Top-right
plt.subplot(2, 2, 2)
plt.plot(sem, cgpa2, marker='o')
plt.title('Year VS Price')
plt.legend(['CGPA2'])
plt.xlabel('Year')
plt.ylabel('Price')

# Bottom-left
plt.subplot(2, 2, 3)
plt.bar(sem, cgpa3)
plt.title('Year VS Price')
plt.legend(['CGPA3'])
plt.xlabel('Year')
plt.ylabel('Price')

# Bottom-right
plt.subplot(2, 2, 4)
plt.bar(sem, cgpa4, width=0.3)
plt.title('Year VS Price')
plt.legend(['CGPA4'])
plt.xlabel('Year')
plt.ylabel('Price')
# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
# Different line styles and markers
plt.figure(figsize=(8, 6))

plt.plot(sem, cgpa1, 'r--*', label='cgpa1')    # Red dashed line with star markers
plt.plot(sem, cgpa2, 'y-^',  label='cgpa2')    # Yellow solid line with triangle markers
plt.plot(sem, cgpa3, 'g-',   label='cgpa3')    # Green solid line

plt.title('Semester VS CGPA')
plt.xlabel('Semester')
plt.ylabel('CGPA')
plt.xlim(0, 9)# X-axis range
plt.ylim(0, 10)# Y-axis range

plt.legend()
plt.show()
```
