#Simple and Multiple Linear Regression using Gradient descent method without using sklearn.

```python
#simple linear regression using batch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Salary_dataset.csv")
print(data.head())

X = data["YearsExperience"].values
y = data["Salary"].values

X = X.reshape(-1, 1)

X_mean = np.mean(X)
X_std = np.std(X)
X = (X - X_mean) / X_std   # Standardization

X_b = np.c_[np.ones((X.shape[0], 1)), X]   # shape (n, 2)


def gradient_descent(X, y, learning_rate=0.1, n_iterations=1000):
    m = len(y)                   # number of samples
    theta = np.zeros(X.shape[1]) # initialize parameters (intercept + slope)
    cost_history = []

    for i in range(n_iterations):
        # Predictions
        y_pred = X.dot(theta)

        # Errors
        error = y_pred - y

        # Gradient calculation
        gradients = (1/m) * X.T.dot(error)

        # Update rule
        theta = theta - learning_rate * gradients

        # Cost (MSE)
        cost = (1/(2*m)) * np.sum(error ** 2)
        cost_history.append(cost)

    return theta, cost_history

theta_best, cost_history = gradient_descent(X_b, y, learning_rate=0.1, n_iterations=1000)

print("Learned Parameters (Intercept, Slope):", theta_best)


y_pred = X_b.dot(theta_best)

slope_real = theta_best[1] / X_std
intercept_real = theta_best[0] - slope_real * X_mean
print("Intercept (real scale):", intercept_real)
print("Slope (real scale):", slope_real)



```

```python
#MULTIPLE LINEAR REGRESSION USING STOCASTIC GD
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Student_Performance.csv")
print(data.head())

# Yes = 1, No = 0
data["Extracurricular Activities"] = data["Extracurricular Activities"].map({"Yes": 1, "No": 0})

# Features (X) and Target (y)
X = data[["Hours Studied", "Previous Scores", "Extracurricular Activities",
          "Sleep Hours", "Sample Question Papers Practiced"]].values
y = data["Performance Index"].values

# Normalize features (important for SGD stability)
X_mean, X_std = X.mean(axis=0), X.std(axis=0)
X = (X - X_mean) / X_std

# Add intercept column
X_b = np.c_[np.ones((X.shape[0], 1)), X]  # shape: (n_samples, n_features+1)

def stochastic_gradient_descent(X, y, learning_rate=0.01, n_epochs=50):
    m, n = X.shape
    theta = np.zeros(n)  # initialize parameters
    cost_history = []

    for epoch in range(n_epochs):
        for i in range(m):
            random_index = np.random.randint(m)   # pick one sample
            xi = X[random_index:random_index+1]   # shape (1, n)
            yi = y[random_index]

            # Prediction
            y_pred = xi.dot(theta)

            # Error
            error = y_pred - yi

            # Gradient
            gradients = xi.T * error

            # Update rule
            theta = theta - learning_rate * gradients.ravel()

        # Compute cost (MSE) after each epoch
        y_pred_all = X.dot(theta)
        cost = (1/(2*m)) * np.sum((y_pred_all - y) ** 2)
        cost_history.append(cost)

    return theta, cost_history

theta_best, cost_history = stochastic_gradient_descent(X_b, y, learning_rate=0.01, n_epochs=100)

print("Learned Parameters (Intercept + Coefficients):", theta_best)

# Example student: 8 hours studied, 90 prev score, extracurricular=1, sleep=7, papers=4
sample = np.array([1, (8-X_mean[0])/X_std[0], (90-X_mean[1])/X_std[1],
                   (1-X_mean[2])/X_std[2], (7-X_mean[3])/X_std[3], (4-X_mean[4])/X_std[4]])
pred_score = sample.dot(theta_best)
print(f"Predicted Performance Index for given student: {pred_score}")
```

```python
#MULTIPLE LINEAR REGRESSION USING MINI BATCH GD
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv("Student_Performance.csv")

# Convert categorical column to numbers (Yes=1, No=0)
data["Extracurricular Activities"] = data["Extracurricular Activities"].map({"Yes": 1, "No": 0})

# Features (X) and Target (y)
X = data[["Hours Studied", "Previous Scores", "Extracurricular Activities",
          "Sleep Hours", "Sample Question Papers Practiced"]].values
y = data["Performance Index"].values

# Add intercept column (for bias term)
X_b = np.c_[np.ones((X.shape[0], 1)), X]   # shape: (rows, features+1)


def mini_batch_gd(X, y, learning_rate=0.0001, epochs=100, batch_size=10):
    m, n = X.shape
    theta = np.zeros(n)   # initialize parameters
    cost_history = []

    for epoch in range(epochs):
        # Shuffle data
        indices = np.random.permutation(m)
        X_shuffled = X[indices]
        y_shuffled = y[indices]

        # Go through batches
        for i in range(0, m, batch_size):
            xi = X_shuffled[i:i+batch_size]
            yi = y_shuffled[i:i+batch_size]

            # Predictions
            y_pred = xi.dot(theta)

            # Gradient calculation
            error = y_pred - yi
            gradients = (1/batch_size) * xi.T.dot(error)

            # Update rule
            theta = theta - learning_rate * gradients

        # Save cost after each epoch
        total_pred = X.dot(theta)
        cost = (1/(2*m)) * np.sum((total_pred - y)**2)
        cost_history.append(cost)

    return theta, cost_history

theta_best, cost_history = mini_batch_gd(X_b, y, learning_rate=0.0001, epochs=200, batch_size=10)

print("Best Parameters (Intercept + Coefficients):")
print(theta_best)

# Example: Student -> 8 hrs study, 85 prev score, extracurricular=1, 7 sleep hrs, 4 papers
sample = np.array([1, 8, 85, 1, 7, 4])   # 1 for intercept
pred_score = sample.dot(theta_best)
print("Predicted Performance Index:", pred_score)

```
