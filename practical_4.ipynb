```python
import pandas as pd
url='https://tinyurl.com/yx3b6sq3'
# Reading CSV file with custom separator and column names
data=pd.read_csv(url,sep=';',header=None,prefix='Column')
# Filtering rows based on multiple conditions
print(data[(data['Column1']==0.580) & (data['Column4']==1)])
```

```python
# Replacing values in a column (without changing the original DataFrame)
d1=data["pclass"].replace("1st","First")
print(data)
print(d1)

# Replacing multiple values with new labels
d2=data["sex"].replace(["female","male"],["woman ","man"]).head()
print(data)
print(d2)
```

```python
# Replacing values directly in the DataFrame (inplace = True)
data['sex'].replace(['female','male'],['Woman','Man'],inplace=True)
print(data)
```

```python
data.rename(columns={'sex':'Gender'},inplace=True)
print(data.head())
```

```python
# Calculating statistical measures on the 'age' column
print(data['age'].mean())# Mean
print(data['age'].var())# Variance
print(data['age'].std())# Standard Deviation
print(data['age'].kurt())# Kurtosis
print(data['age'].skew())# Skewness
print(data['age'].sem())# Standard Error of Mean
print(data['age'].mode())# Mode
print(data['age'].median())# Median
```

```python
# Getting unique values of a column
print(data['pclass'].unique)
# Counting frequency of unique values
print(data['age'].value_counts())
print(data['pclass'].value_counts())
```

```python
# Dropping a single column (age) without modifying original data
d1=data.drop('age',axis=1)
import numpy as np
# Setting display options for better readability
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
print(d1.head())# DataFrame after drop
print(data.head())
# Dropping column 'age' from original DataFrame (inplace)
data.drop('age',axis=1,inplace=True)
```

```python
# Dropping multiple columns at once
d1=data.drop(['parch','pclass'],axis=1).head()
print(d1)
```

```python
# Dropping a column by index (2nd column, index=1)
d1=data.drop(data.columns[1],axis=1).head()
print(d1)
```

```python
#deleting rows
# Filtering rows where fare > 30
print(data[data["fare"]>30])
d1=data[data["fare"]>30]
print(d1)
```

```python
# Dropping duplicate rows
print(data.drop_duplicates())
print(data)
d1=data.drop_duplicates()
print(d1)
# Dropping duplicate rows based on a specific column ('fare')
d2=data.drop_duplicates(subset=['fare'])
print(d2)
```

```python
# Grouping by 'survived' column and counting occurrences
print(data.groupby('survived').count())
```
