```python
#Using different separator from file like ; or other
# Read CSV file with semicolon as separator
data = pd.read_csv("employees.csv", sep=',')
print(data)
```

```python
# Using column names explicitly
data = pd.read_csv("employees.csv", names=["EmpID","FullName","Dept","JoinDate","Pay"])
print(data)
```

```python
# Using any column as index
data = pd.read_csv("employees.csv", index_col=0)  # ID as index
print(data)
```

```python
# Getting selected columns by index
data = pd.read_csv("employees.csv", usecols=[1,2,4])  # Name, Dept, Salary
print(data)
```

```python
# Read CSV with no header and assign custom names
data = pd.read_csv("employees.csv", header=None, names=[f"Column{i}" for i in range(5)])
print(data)
```

```python
# Skipping selected rows (e.g., skip odd rows)
data = pd.read_csv("employees.csv", skiprows=lambda x: x % 2 == 0)
print(data)
```
